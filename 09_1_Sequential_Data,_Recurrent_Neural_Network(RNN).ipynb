# 순차 데이터

      텍스트나 시계열 데이터(time series data)와 같이 순서에 의미가 있는 데이터

      * 순차 데이터는 순서가 유의미하므로 테스트, 검증 세트의 구분에 있어 랜덤셔플에 유의하여야 함

## 피드포워드 신경망
      완전 연결 신경망과 합성곱 신경망과 같이 하나의 샘플을 사용하여 정방향 계산을 수행하고 나면 해당 샘플은 다음 샘플의 처리에 재사용되지 않는 방식의 신경망

 # 순환 신경망  Recurrent Neural Network, RNN
      다음 샘플의 데이터 처리 과정에 이전 데이터가 신경망 층에 순환하는 구조

## 타임스텝 timestep
      샘플을 처리하는 하나의 단계


## 셀 cell
      한 층에 있는 뉴런의 집합

## 은닉 상태 hidden state
      셀의 출력

## 순환 신경망 = 완전 연결 신경망의 타임 스텝에 셀의 은닉 상태를 추가한 구조
        순환 신경망에서는 셀에 포함된 뉴런의 각 출력(은닉 상태)이 회귀하여 다음 데이터 처리(타임 스텝)에 포함됨

## 활성화 함수
      하이퍼볼릭 탄젠드 hyperbolic tngent 함수인 tanh가 많이 사용
      시그모이드 함수와 유사하지만 -1 ~ 1의 범위를 가진다는 점에서 차이를 둠
      순환 신경망 역시 반드시 활성화 함수가 필요함

## 순환 신경망에서의 가중치
      입력에 곱해지는 가중치 Wx - 피드포워드 신경망들과 동일
      **이전 타임스텝의 은닉 상태에 곱해지는 가중치 Wb가 추가됨**

ㅋㅋ.jpg

### 가중치 크기(개수)
      가정) 순환층에 입력되는 특성의 개수 4, 순환층의 뉴런이 3

      
      

1.jpg

      4 X 3 = 12
      
      Wx = 12개

2.jpg

      각 뉴런이 자신에게 회귀, 다른 모든 뉴런과 완전 연결되어 있으므로 
      3 X 3 = 9
      Wb = 9

## 모델 파라미터 개수
      Wx + Wb + 절편 = 12 + 9 + 3 (뉴런 개수) = 24

## 시퀀스 sequence
      순환 신경망에서 하나의 샘플
      시퀀스의 길이는 타임스텝의 길이가 됨. 텍스트의 경우 단어의 수가 됨

## 순환 신경망의 샘플의 크기
      일반적으로 2개의 차원을 가짐, (시퀀스 길이(타입스텝), 단어 표현)
